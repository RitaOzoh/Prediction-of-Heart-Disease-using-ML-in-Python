## MACHINE LEARNING

#### IMPORT THE REQUIRED LIBRARIES FOR MACHINE LEARNING

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, plot_confusion_matrix, classification_report, confusion_matrix, auc,roc_curve
```

#### SPLIT THE DATASET

```python
x=heart.drop("HeartDiseaseorAttack", axis=1)
y=heart.HeartDiseaseorAttack
```

#### NORMALIZE THE DATASET

```python
#Normalize the value of x
scaler = MinMaxScaler()
scaler.fit(x)
norm_heart = scaler.transform(x)
norm_heart = pd.DataFrame(norm_heart, columns=x.columns)
norm_heart.head()
```

#### SPLIT INTO TEST AND TRAIN

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 0)

print(x_train.shape), print(y_train.shape)
print(x_test.shape), print(y_test.shape)
```

```python
y_train.value_counts()
```

#### IMPORT LIBRARIES FOR BALANCING DATASET

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

#### BALANCING OF THE DATASET

```python
#Undersampling the train dataset
undersample = RandomUnderSampler(sampling_strategy=0.6)
steps = [('u', undersample)]
pipeline = Pipeline(steps=steps)

x_train, y_train = pipeline.fit_resample(x_train, y_train)
```

```python
y_train.value_counts()
```

```python
plt.figure(figsize = (20,10))
label=['No Heart Disease','Heart Disease']
value_label=y_train.value_counts()
explode=[0.03,0.03]
colors = ['#66b3ff','#99ff99']
fig,ax=plt.subplots()
ax.pie(value_label,explode=explode,labels=label, autopct='%1.2f%%', colors = colors, shadow=True, startangle=90)
ax.set_title('Distribution of predicted values after undersampling', fontdict = {"fontsize":20}, pad = 12)
```

#### BUILDING MACHINE LEARNING MODELS

```python
Models = ["Neural_Network", "KNN", "Gradient_Boosting", "logistic_Regression",
         "Random_Forest", "XGB", "Naive_Bayes", "QDA", "SGD", "Linear_SVM", "Decision_Tree", "AdaBoost"]


classifiers = [
    MLPClassifier(alpha=1, max_iter=1000),
    KNeighborsClassifier(5),
    GradientBoostingClassifier(n_estimators=100, learning_rate=1.0),
    LogisticRegression(max_iter = 1000),
    RandomForestClassifier(max_depth=5, n_estimators=100),
    XGBClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
    SGDClassifier(loss="hinge", penalty="l2"),
    SVC(kernel="linear", C=0.025),
    DecisionTreeClassifier(max_depth=5),
    AdaBoostClassifier(n_estimators=100)]
```

#### PREDICTIONS FROM THE MACHINE LEARNING MODELS

```python
    scores = {}
rita_code = {}
for model in classifiers:
    model.fit(x_train, y_train)
    rita_code[model] = model
    y_pred = model.predict(x_test)
    scores["Accuracy"] = scores.get("Accuracy", []) + [accuracy_score(y_test, y_pred)]
    scores["Precision"] = scores.get("Precision", []) + [precision_score(y_test, y_pred)]
    scores["Recall"] = scores.get("Recall", []) + [recall_score(y_test, y_pred)]
    scores["F1"] = scores.get("F1", []) + [f1_score(y_test, y_pred)]
    scores["AUC"] = scores.get("AUC", []) + [roc_auc_score(y_test, y_pred)]
```

#### RESULTS OF THE PREDICTIONS

```python
results = pd.DataFrame(data=scores, index=Models)
results
```

#### RESULTS OF THE PREDICTIONS IN DESCENDING ORDER

```python
results.sort_values(by= ["Accuracy","F1"], ascending=False)
```

#### PLOTTING CONFUSION MATRIX

```python
_,ax = plt.subplots(4,3,figsize=(10,13))
i=0
for key,val in rita_code.items():   
  #Getting the predicted y model
  y_predict = val.predict(x_test)

  #Plotting the confusion matrix
  conf_matrix = confusion_matrix(y_test, y_predict)
  ax[i//3,i%3].set_title(key,size=8,weight=1000)
  sns.heatmap(pd.DataFrame(conf_matrix,columns=['True','False'],index=['True','False']),annot=True,cbar=False,fmt='.0f', ax=ax[i//3,i%3],cmap=plt.cm.ocean_r)
  ax[i//3,i%3].set_ylabel(' ')
  ax[i//3,i%3].set_xlabel(' ')
  i=i+1
plt.show()  
```

#### ROC CURVE OF THE MODELS

```python
plt.figure(figsize=(15,10))
c=['#a83285','#5b32a8','#5b32a8','#5b32a8','#5b32a8','#89a832','#a89832','#a87532','#329ca8', '#c10b27', '#ff66c4', '#ff4900']
i=0
plt.plot([0,1],[0,1],linewidth=2,c='black', linestyle='--')

for key,val in rita_code.items():
  y_predict = val.predict(x_test)
  fpr,tpr,_ = roc_curve(y_test,y_predict)
  plt.plot(fpr,tpr,linewidth=2,label=str(key) + ':'+ ' %.2f '%auc(fpr,tpr),c=c[i])
  plt.suptitle('Receiver Operation Characteristics')
  plt.xlabel('False Positive')
  plt.ylabel('True Positive')
  plt.legend(loc=0)
  i+=1
plt.show()
```
